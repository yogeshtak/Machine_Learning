### Matplotlib example workflow

```python
#import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x = [1,2,3,4,5]
y = [11,22,56,66,78]

#setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width and height

#plot data
ax.plot(x,y)

#customize plot
ax.set(title= "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#save and show plot
fig.savefig("images/sample-plot.png")
```

```python
#option2
fig, ax = plt.subplots(nrows =2,
                      ncols =2,
                      figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

**which one to use (pyplot vs matplotlib OO method)**
* When plotting something quickly, okay to use the pyplot method
* when plotting something advanced, use the OO (Object oriented) matplotlib method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter", 
             x = "age",
             y = 'chol',
             c = 'target')
```

```python
#matplotlib OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x = 'age',
            y ='chol',
            c = 'target',
            ax= ax)

ax.set_xlim([45,100])
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

#customize the plot
ax.set(title="heart Disease and Cholestrol Levels",
      xlabel = "Age",
      ylabel = "Cholestrol")

#add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = '--')
```

```python
over_50.head()
```

```python
#subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10))
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel ="age",
       ylabel="Cholestrol")

#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

#customise ax1
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel = "age",
       ylabel = "Thalach")

#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")
```

```python
#adding shared x-axis and figure title

#subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10),
                             sharex=True)  #shared axis
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

#customise ax1
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel = "age",
       ylabel = "Thalach")

#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

#add a title for figure, whole figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold")
```

```python
plt.style.use('fivethirtyeight')
#adding shared x-axis and figure title

#subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10),
                             sharex=True)  #shared axis
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")  #this changes color scheme

#customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                      cmap="plasma")  #this changes color scheme

#customise ax1
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel = "age",
       ylabel = "Thalach")

#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

#add a title for figure, whole figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold")
```

```python
#customising x and y axis limitations

plt.style.use('tableau-colorblind10')
#adding shared x-axis and figure title

#subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10),
                             sharex=True)  #shared axis
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")  #this changes color scheme

#customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

#change axis limite
ax0.set_xlim([50,80])

#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                      cmap="winter")  #this changes color scheme

#customise ax1
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel = "age",
       ylabel = "Thalach")

#change x1 axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

#add a title for figure, whole figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold")
```
