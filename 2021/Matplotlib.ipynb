### Matplotlib example workflow

```python
#import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x = [1,2,3,4,5]
y = [11,22,56,66,78]

#setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width and height

#plot data
ax.plot(x,y)

#customize plot
ax.set(title= "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#save and show plot
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

* Line Plot
* Scatter Plot
* Bar Plot
* Histogram

```python
#create some data
x = np.linspace(0,10,100)
x
```

```python
#plot the data
fig,ax = plt.subplots()
ax.plot(x,x**2)
```

```python
#scatter
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
#make plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 6,
                    "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter Store",
      ylabel= "Price" )
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
#make some data
#randn is for normal distribution
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
# subplots option 1

fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#option2
fig, ax = plt.subplots(nrows =2,
                      ncols =2,
                      figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```
